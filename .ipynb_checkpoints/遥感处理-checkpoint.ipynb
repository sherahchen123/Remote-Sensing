```python
# 
dataset = gdal.Open('F:\\\\GIS\\LC09_L2SP_034032_20240806_20240807_02_T1_SR_B4.tif', gdal.GA_Update)
# ...
image_data = dataset.ReadAsArray()
# 
width = dataset.RasterXSize
height = dataset.RasterYSize
 
# matplotlib
plt.imshow(image_data, cmap='gray')
plt.axis('off')  # 
plt.show()
dataset.GetMetadata()  # 
dataset.GetDescription()  #  
dataset.RasterCount  # 
# MODISL1B 0  dataset  0  MODISL1BHDF
# 
# dataset = None
```

```python
# ReadAsArray()(Array)
data_array = dataset.ReadAsArray()
print(data_array.shape)
band1 = dataset.GetRasterBand(1)  # 
band1_array = band1.ReadAsArray()
print(band1_array.shape)
band1.XSize
band1.YSize
band1.ReadAsArray(100,100,10,10)
```

```python
# ReadAsArray()
# xoffyoff
# xsizeysize
data_array = dataset.ReadAsArray(xoff=0, yoff=0, xsize=20, ysize=15)
print(data_array)
print(data_array.shape)
```

```python
help(dataset.ReadRaster)  
# help
```

```python
import numpy
band2 = dataset.GetRasterBand(2)
band3 = dataset.GetRasterBand(3)
cols = 100
rows = 100
data2 = band2.ReadAsArray(0, 0, cols,rows).astype(numpy.float16)
data3 = band3.ReadAsArray(0, 0, cols,rows).astype(numpy.float16)
mask = numpy.greater(data3 + data2, 0)
# NDVI=(NIRRED)/(NIR+RED)  NIR:band3; RED:band2;
ndvi = numpy.choose(mask, (-99, (data3 - data2) / (data3 + data2)))
print(ndvi)
```
