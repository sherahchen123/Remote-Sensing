```python
from osgeo import gdal
import matplotlib.pyplot as plt
```

```python
# 打开数据集以进行更新
dataset = gdal.Open('F:\\数据分析测试数据\\GIS遥感数据\\LC09_L2SP_034032_20240806_20240807_02_T1_SR_B4.tif', gdal.GA_Update)
# 对数据集进行修改等操作...
image_data = dataset.ReadAsArray()
# 获取图像的宽度和高度
width = dataset.RasterXSize
height = dataset.RasterYSize
 
# 使用matplotlib显示图像
plt.imshow(image_data, cmap='gray')
plt.axis('off')  # 不显示坐标轴
plt.show()
dataset.GetMetadata()  # 获取元数据
dataset.GetDescription()  # 获取栅格的描述 图像的路径名
dataset.RasterCount  # 几个波段组成的影像
# MODISL1B数据的话，运行结果居然是 0 。这意味着当前的数据集 dataset 中的栅格数目是 0 。 实际上，MODISL1B的数据格式是HDF格式的，它的数据是以子数据集组织的，要获取其相关的数据的信息，需要继续访问其子数据集。
# 关闭数据集
# dataset = None
```

```python
rows = dataset.RasterYSize
cols = dataset.RasterXSize
bands = dataset.RasterCount
print(f"rows:{rows}")
print(f"cols:{cols}")
print(f"bands:{bands}")
```

```python
# 返回栅格数据的坐标转换参数，即行列坐标与空间坐标的转换参数
geotrans = dataset.GetGeoTransform()
print(geotrans)
```

```python
# 返回栅格数据的坐标转换参数
proj = dataset.GetProjection()
print(proj)
```

```python
# ReadAsArray()方法可实现将栅格数据转换成数组(Array)形式，以便后续处理
data_array = dataset.ReadAsArray()
print(data_array.shape)
band1 = dataset.GetRasterBand(1)  # 获取第一个波段的数据
band1_array = band1.ReadAsArray()
print(band1_array.shape)
band1.XSize
band1.YSize
band1.ReadAsArray(100,100,10,10)
```

```python
# ReadAsArray()同样支持按块读取栅格信息，即读取部分区域图像信息
# xoff，yoff：想要读取的部分原点位置在整张图像中距离全图原点的位置
# xsize和ysize指定要读取部分图像的矩形大小
data_array = dataset.ReadAsArray(xoff=0, yoff=0, xsize=20, ysize=15)
print(data_array)
print(data_array.shape)
```

```python
help(dataset.ReadRaster)  
# 用help方法查看方法的使用说明
```

```python
import numpy
band2 = dataset.GetRasterBand(2)
band3 = dataset.GetRasterBand(3)
cols = 100
rows = 100
data2 = band2.ReadAsArray(0, 0, cols,rows).astype(numpy.float16)
data3 = band3.ReadAsArray(0, 0, cols,rows).astype(numpy.float16)
mask = numpy.greater(data3 + data2, 0)
# NDVI = (NIR− RED)/(NIR+RED)  NIR:band3; RED:band2;
ndvi = numpy.choose(mask, (-99, (data3 - data2) / (data3 + data2)))
print(ndvi)
```

```python
# DataType是指图像中实际数值的数据类型。具体数据类型定义在gdalconst模块里。
from osgeo import gdalconst
# 查看某一波段的数据类型
print(band1.DataType)
# dir 获得当前模块的属性列表
print(dir(gdalconst)[:6] + ['... ...'] + dir(gdalconst)[-3:])
# 未知或未指定类型 gdalconst.GDT_Unknown 0
# 8位无符整型 gdalconst.GDT_Byte 1
# 16位无符整型 gdalconst.GDT_UInt16 2
# 16位整型 gdalconst.GDT_Int16 3
# 32位无符整型 gdalconst.GDT_UInt32 4
# 32位整型值 gdalconst.GDT_Int32 5
# 32位浮点型 gdalconst.GDT_Float32 6
# 64位浮点型 gdalconst.GDT_Float64 7
# 16位复数整型 gdalconst.GDT_CInt16 8
# 32位复数整型 gdalconst.GDT_CInt32 9
# 32位复数浮点型 gdalconst.GDT_CFloat32 10
# 64位复数浮点型 gdalconst.GDT_CFloat64 11
dir(dataset)[:3] + ['... ...'] + dir(dataset)[-3:]
```
